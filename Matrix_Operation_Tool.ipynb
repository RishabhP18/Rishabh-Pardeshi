```python
from pyngrok import ngrok

!pip install streamlit numpy pyngrok

%%writefile matrix_tool.py
import streamlit as st
import numpy as np

st.set_page_config(page_title="Matrix Operations Tool", layout="centered")
st.title("ğŸ§® Matrix Operations Tool with NumPy")

def parse_matrix(text):
    try:
        lines = text.strip().split('\n')
        matrix = [list(map(float, row.strip().split())) for row in lines]
        return np.array(matrix)
    except Exception as e:
        st.error("âš ï¸ Invalid matrix format. Use space-separated values. Example:\n1 2\n3 4")
        return None

col1, col2 = st.columns(2)

with col1:
    st.subheader("Matrix A")
    matrix_a_text = st.text_area("Enter Matrix A", "1 2\n3 4", height=120)

with col2:
    st.subheader("Matrix B")
    matrix_b_text = st.text_area("Enter Matrix B", "5 6\n7 8", height=120)

matrix_a = parse_matrix(matrix_a_text)
matrix_b = parse_matrix(matrix_b_text)

st.markdown("---")
st.subheader("Select Operation")

operation = st.selectbox("Choose operation", [
    "Addition (A + B)",
    "Subtraction (A - B)",
    "Multiplication (A Ã— B)",
    "Transpose of A",
    "Transpose of B",
    "Determinant of A",
    "Determinant of B"
])

if st.button("ğŸ” Calculate"):
    if operation in ["Addition (A + B)", "Subtraction (A - B)", "Multiplication (A Ã— B)"]:
        if matrix_a is None or matrix_b is None:
            st.warning("Please enter both matrices.")
        else:
            try:
                if operation == "Addition (A + B)":
                    result = matrix_a + matrix_b
                elif operation == "Subtraction (A - B)":
                    result = matrix_a - matrix_b
                elif operation == "Multiplication (A Ã— B)":
                    result = np.dot(matrix_a, matrix_b)
                st.success("âœ… Result:")
                st.write(result)
            except Exception as e:
                st.error(f"âŒ Error: {e}")

    elif operation == "Transpose of A":
        if matrix_a is not None:
            st.success("âœ… Transpose of A:")
            st.write(matrix_a.T)
        else:
            st.warning("Please enter Matrix A.")

    elif operation == "Transpose of B":
        if matrix_b is not None:
            st.success("âœ… Transpose of B:")
            st.write(matrix_b.T)
        else:
            st.warning("Please enter Matrix B.")

    elif operation == "Determinant of A":
        if matrix_a is not None:
            if matrix_a.shape[0] == matrix_a.shape[1]:
                det = np.linalg.det(matrix_a)
                st.success(f"âœ… Determinant of A: {round(det, 4)}")
            else:
                st.warning("Matrix A must be square to compute determinant.")
        else:
            st.warning("Please enter Matrix A.")

    elif operation == "Determinant of B":
        if matrix_b is not None:
            if matrix_b.shape[0] == matrix_b.shape[1]:
                det = np.linalg.det(matrix_b)
                st.success(f"âœ… Determinant of B: {round(det, 4)}")
            else:
                st.warning("Matrix B must be square to compute determinant.")
        else:
            st.warning("Please enter Matrix B.")

ngrok.kill()
public_url = ngrok.connect(port=8501)
print("Public URL:", public_url)

!streamlit run matrix_tool.py --server.port 8501 --server.enableCORS false
```
